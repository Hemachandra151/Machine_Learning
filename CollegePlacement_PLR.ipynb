#Train Test Split

```python
x,tx=train_test_split(data,random_state=2529)
farm=x[['Placement']]
farm=farm.values.tolist()
key=['Yes','No']
```

#Activation Function

```python
def sig(a):
  a*=-1
  theta=1/(1+math.exp(a))
  if theta<0.5:
    return 0
  else:
    return 1
```

#Perceptron Logistic Regression

```python
col=['Python', 'DeepLearning', 'MachineLearning', 'BigData', 'CloudComputing']
setv=x[col]
setv=setv.values.tolist()

w=np.zeros(6)
eta=0.2

for i in range(5):
  w[i]=rn.randint(-10,10)
w[5]=1

for v in setv:
  v.append(1)

iter=0
while iter<400:
  j=0
  for v in setv:
    norm=np.dot(w,v)
    y=sig(norm)

    for i in range(6):
      w[i]=w[i]-eta*(y-key.index(farm[j][0]))*v[i]

    j+=1
  iter+=1

def plr(v):
  norm=np.dot(w,v)
  return sig(norm)
```

#Testing The Model

```python
N=len(tx)
para=tx[col]
para=para.values.tolist()
out=tx[['Placement']]
out=out.values.tolist()
n=0
for i in range(N):
  para[i].append(1)
  if plr(para[i])==key.index(out[i][0]):
    n+=1
print(n/N*100)
print(n)
print(N)
```
