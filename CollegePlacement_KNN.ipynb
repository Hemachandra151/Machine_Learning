#Train Test Split

```python
x,tx=train_test_split(data,random_state=2529)
farm=x.values.tolist()
key=['Yes','No']
```

#K-NN Classifier

```python
def knn(py,dl,ml,bd,cc):
  k=15
  n=len(farm)
  d=np.full(k,np.nan)
  norm=np.zeros(n)
  freq=np.zeros(2)

  for i in range(n):
    pyd=farm[i][0]-py
    dld=farm[i][1]-dl
    mld=farm[i][2]-ml
    bdd=farm[i][3]-bd
    ccd=farm[i][4]-cc

    norm[i]=math.sqrt(pyd*pyd+dld*dld+mld*mld+bdd*bdd+ccd*ccd)

  count=0
  while count<k:
    mini=9999
    for i in range(n):
      if norm[i]<mini and i not in d:
        mini=norm[i]
        val=i
    d[count]=val
    count+=1

  for i in range(k):
    freq[key.index(farm[int(d[i])][5])]+=1


  return key[np.argmax(freq)]
```

#Testing The Model

```python
N=len(tx)
para=tx.values.tolist()
n=0
for i in range(N):
  if knn(para[i][0],para[i][1],para[i][2],para[i][3],para[i][4])==para[i][5]:
    n+=1

print(n/N*100)
print(n)
print(N)
```
