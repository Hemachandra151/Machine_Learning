#Train Test Split

```python
x,tx = train_test_split(data, random_state=2529)
farm = x[['Default']].values.tolist()
```

#Activation Function

```python
def lay1(a):
  a*=-1
  return 1/(1+np.exp(a))

def lay2(a):
  a*=-1
  t=1/(1+np.exp(a))
  if t<0.5:
    return 0
  else:
    return 1
```

#Multilayer Perceptron

```python
col=['Income', 'Age', 'Loan', 'Loan to Income']
att = x[col].values.tolist()

for i in att:
  i.append(1)

#Layer 1 Perceptron 1
w1=np.array([0.01,-0.05,0.04,0.07,-0.01])
#Layer 1 Perceptron 2
w2=np.array([-0.01,0.02,-0.01,0.03,0.01])

#Layer 2 Perceptron 1
v1=np.array([0.01,-0.01,0.01])


iter=0
miter=1
eta = 0.02
while iter<miter:
  for i,V in enumerate(att):
    v = np.array(V)
    #Forward Phase
    a1 = np.dot(v,w1)
    z1 = lay1(a1)
    a2 = np.dot(v,w2)
    z2 = lay1(a2)

    z =np.array([z1,z2,1])
    ao = np.dot(v1,z)
    y = lay1(ao)

    #Backward Phase
    errAtOut = (farm[i][0] - y)*(y*(1-y))*ao
    errAtHL = np.zeros(3)
    for c in range(2):
      errAtHL[c] = (y*(1-y))*(ao)*(errAtOut)*v1[c]

    #Update Layer 2 Weights
    for c in range(3):
      v1[c] = v1[c]+eta*errAtOut*z[c]

    #Update Layer 1 Weights
    for c in range(5):
      w1[c] = w1[c]+eta*errAtHL[0]*v[c]
      w2[c] = w2[c]+eta*errAtHL[1]*v[c]

  iter += 1

print(f"{w1} {w2} {v1}")
def mlp(v):
  z1 = lay1(np.dot(w1,v))
  z2 = lay1(np.dot(w2,v))

  z = np.array([z1,z2,1])

  a = np.dot(z,v1)
  return lay2(a)

```

#Testing The Model

```python
N = len(tx)
para = tx[col].values.tolist()
out = tx[['Default']].values.tolist()
conf=np.zeros((2,2))

for i,v in enumerate(para):
  v.append(1)
  if mlp(v)==out[i][0]:
    if out[i][0]==0:
      conf[0][0]+=1
    else:
      conf[1][1]+=1
  else:
    if out[i][0]==0:
      conf[0][1]+=1
    else:
      conf[1][0]+=1
n=conf[0][0]+conf[1][1]
N=n+conf[0][1]+conf[1][0]
print(n/N*100)
print(f"Out of {conf[0][0]+conf[0][1]} targets labled as 0, {conf[0][0]} were classified correctly")
print(f"Out of {conf[1][1]+conf[1][0]} targets labled as 1, {conf[1][1]} were classified correctly ")
print(n)
print(N)
```
