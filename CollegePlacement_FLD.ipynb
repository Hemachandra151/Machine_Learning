#Train Test Split

```python
x,tx=train_test_split(data,random_state=2529)
key=['Yes','No']
```

#Fisher's Linear Discriminant

```python
col=['Python', 'DeepLearning', 'MachineLearning', 'BigData', 'CloudComputing']
D1=x[x['Placement']=='Yes']
D2=x[x['Placement']=='No']
D1=D1[col]
D2=D2[col]
D1=D1.values.tolist()
D2=D2.values.tolist()

md1=np.zeros(5)
md2=np.zeros(5)

for v in D1:
  for i in range(5):
    md1[i]+=v[i]
md1/=len(D1)

for v in D2:
  for i in range(5):
    md2[i]+=v[i]
md2/=len(D2)

sd1=np.zeros((5,5))
for v in D1:
  temp=np.zeros((5,5))
  vec=(v-md1)
  for i in range(5):
    for j in range(5):
      temp[j][i]=vec[j]*vec[i]
  for i in range(5):
    for j in range(5):
      sd1[i][j]=sd1[i][j]+temp[i][j]

sd2=np.zeros((5,5))
for v in D2:
  temp=np.zeros((5,5))
  vec=(v-md2)
  for i in range(5):
    for j in range(5):
      temp[j][i]=vec[j]*vec[i]
  for i in range(5):
    for j in range(5):
      sd2[i][j]=sd2[i][j]+temp[i][j]

cvm=(1/(len(D1)+len(D2)-2))*(sd1+sd2)

W=np.linalg.inv(cvm) @ (md1-md2)

m1=0
for v in D1:
  m1+=np.dot(W,v)
m1/=len(D1)
print(m1)

m2=0
for v in D2:
  m2+=np.dot(W,v)
m2/=len(D2)
print(m2)

print(len(D1))
print(len(D2))
print(len(x))

t=(m1+m2)/5
print(t)

def fld(ve):
  norm=np.dot(W,ve)
  if norm>=t:
    return 'Yes'
  else:
    return 'No'
```

#Testing The Model

```python
N=len(tx)
para=tx.values.tolist()
ve=tx[['Python', 'DeepLearning', 'MachineLearning', 'BigData', 'CloudComputing']]
ve=ve.values.tolist()
n=0
for i in range(N):
  if fld(ve[i])==para[i][5]:
    n+=1

print(n/N*100)
print(n)
print(N)
```
