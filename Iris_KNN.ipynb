#Train Test Split

```python
x,tx=train_test_split(data,random_state=2529)
key=['Iris-setosa','Iris-versicolor','Iris-virginica']
farm=x.values.tolist()
```

#K-NN Classifier

```python
def knn(sl,sw,pl,pw):
  k=15
  n=len(farm)
  d=np.full(k,np.nan)
  norm=np.zeros(n)
  freq=np.zeros(3)

  for i in range(n):
    sld=farm[i][0]-sl
    swd=farm[i][1]-sw
    pld=farm[i][2]-pl
    pwd=farm[i][3]-pw

    norm[i]=math.sqrt(sld*sld+swd*swd+pld*pld+pwd*pwd)

  count=0
  while count<k:
    mini=9999
    for i in range(n):
      if norm[i]<mini and i not in d:
        mini=norm[i]
        val=i
    d[count]=val
    count+=1

  for i in range(k):
    freq[key.index(farm[int(d[i])][4])]+=1

  return key[np.argmax(freq)]
```

#Testing The Model

```python
N=len(tx)
para=tx.values.tolist()
n=0
for i in range(N):
  if knn(para[i][0],para[i][1],para[i][2],para[i][3])==para[i][4]:
    n+=1

print(n/N*100)
print(n)
print(N)
```
