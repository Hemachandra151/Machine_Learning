#Train Test Split

```python
x,tx=train_test_split(data, random_state=2529)
farm=x[['diabetes']].values.tolist()
```

#Activation Function

```python
def sig(a):
  a*=-1
  theta=1/(1+np.exp(a))
  if theta<0.5:
    return 0
  else:
    return 1
```

#Perceptron Logistic Regression

```python
col=['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi', 'dpf', 'age']
setv=x[col].values.tolist()

w=np.zeros(9)
eta=0.2
rn.seed(32)

for i in range(8):
  w[i]=rn.randint(-5,5)
w[8]=rn.randint(1,3)

for v in setv:
  v.append(1)

iter=0
while iter<250:
  j=0
  for v in setv:
    norm=np.dot(w,v)
    y=sig(norm)

    for i in range(9):
      w[i]=w[i]-eta*(y-farm[j][0])*v[i]

    j+=1
  iter+=1

def plr(v):
  norm=np.dot(w,v)
  return sig(norm)
```

#Testing The Model

```python
N=len(tx)
para=tx[col].values.tolist()
out=tx[['diabetes']].values.tolist()

conf=np.zeros((2,2))
for i in range(N):
  para[i].append(1)
  if plr(para[i])==out[i][0]:
    if out[i][0]==0:
      conf[0][0]+=1
    else:
      conf[1][1]+=1
  else:
    if out[i][0]==0:
      conf[0][1]+=1
    else:
      conf[1][0]+=1

n=conf[0][0]+conf[1][1]
N=n+conf[0][1]+conf[1][0]
print(n/N*100)
print(f"Out of {conf[0][0]+conf[0][1]} targets labled as 0, {conf[0][0]} were classified correctly")
print(f"Out of {conf[1][1]+conf[1][0]} targets labled as 1, {conf[1][1]} were classified correctly ")
print(n)
print(N)
```
