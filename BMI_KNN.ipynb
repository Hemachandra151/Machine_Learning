```python
df.replace({'Gender':{'Male':0,'Female':1}},inplace=True)
```

#Train Test Split

```python
x,tx=train_test_split(df,random_state=2529)
farm=x.values.tolist()
```

# K-NN Classifier

```python
def knn(g,h,w):
  k=11
  n=len(farm)
  d=np.full(k,np.nan)
  norm=np.zeros(n)
  freq=np.zeros(6)

  for i in range(n):
    gd=farm[i][0]-g
    hd=farm[i][1]-h
    wd=farm[i][2]-w

    norm[i]=math.sqrt(gd*gd+hd*hd+wd*wd)

  count=0
  while count<k:
    mini=9999
    for i in range(n):
      if norm[i]<mini and i not in d:
        mini=norm[i]
        val=i
    d[count]=val
    count+=1

  for i in range(k):
    freq[farm[int(d[i])][3]]+=1

  return np.argmax(freq)
```

#Testing The Model

```python
N=len(tx)
para=tx.values.tolist()
n=0
for i in range(N):
  if knn(para[i][0],para[i][1],para[i][2])==para[i][3]:
    n+=1

print(n/N*100)
print(n)
print(N)
```
